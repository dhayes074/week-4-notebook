### FizzBuzz if num is divisible by 3 and 5 or Fizz if 3 and Buzz if 5. Returns num if not divisible.

```python
def fizz_buzz(num):
	return "Fizz"*(num%3==0) + "Buzz"*(num%5==0) or str(num)

for i in [3,5,15,2]:
	print(fizz_buzz(i))
```
